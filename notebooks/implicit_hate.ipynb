```python
import os
import pandas as pd
```

```python
directory = "../data/implicit-hate-corpus"
tsv_files = [f for f in os.listdir(directory) if ".tsv" in f]
dfs = {}
for tsv in tsv_files:
    filepath = os.path.join(directory, tsv)
    df = pd.read_csv(filepath, sep="\t")
    dfs[tsv] = df
```

```python
dfs.keys()
```

```python
for filename, df in dfs.items():
    print(filename, df.shape)
    print(df.columns)
    print()
```
